```python

preds = ['ab', 'abcd', 'dabc', 'cbad', 'adbc', '', 'abcdefg', 'gfdefabc']
actuals = ['abc'] * len(preds)
actual = list('abc')
scores = []

for predicted in preds:
    predicted = list(predicted)
    score = 0.0
    num_hits = 0.0
    print()
    for i,p in enumerate(predicted):
        if p in actual and p not in predicted[:i]:
            num_hits += 1.0
            score += num_hits / (i+1.0)
        print(score)
    scores.append(score / min(len(actual), 12))

pd.DataFrame({"y_true": [list(a) for a in actuals],    
              "y_pred": [list(p) for p in preds],
              "AP@12" : scores})
```
